# Graphical Representation of Data 
This notebook takes the data table created from the previous notebook (bbb_permeability_data_analysis) and uses the computed values to graph a range of histograms and box plots

## Import 
Import the python file needed to graph the data

```python
from data_analysis import * 
import numpy as np, matplotlib.pyplot as plt 
```

## Variables 
Define the variables needed to graph the data

```python
tpsa_values = [tpsa_positive, tpsa_negative]
logP_values = [logP_positive, logP_negative]
```

## Graph a Histogram of Molecular Weights for combined data table
Graph a histogram that shows the overall distribution of molecular weight for the entire data table

```python
'''Make a Histogram of Molecular Weights for combined dataset'''
bins = np.arange(min(molecular_weights), max(molecular_weights) + 1, 50)
plt.hist(molecular_weights, edgecolor='black', align='mid', bins = bins)
ax = plt.gca()
ax.set_xlabel('Molecular Weight (amu)')
ax.set_ylabel('Number of Molecules')
ax.set_title("Number of Molecules in each Molecular Weight (amu) Category", size=11.5, weight='bold') 
plt.show()
```

## Graph a Histogram of LogP Value for combined data table 
Graph a histogram that shows the overall distribution of logP for the entire data table

```python
'''Make a Histogram of LogP Value for combined dataset'''
bins = np.arange(min(logPs), max(logPs) + 1, 1)
plt.hist(logPs, edgecolor ='black', align='mid', bins = bins)
ax = plt.gca()
ax.set_xlabel('LogP Value')
ax.set_ylabel('Number of Molecules')
ax.set_title("Number of Molecules in each LogP Value Category", size=13.5, weight='bold') 
plt.show()
```

## Graph a Histogram of TPSA for combined data table
Graph a histogram that shows the overall distribution of TPSA for the entire data table

```python
'''Make a Histogram of the TPSA for combined dataset'''
plt.hist(logPs, edgecolor ='black', align='mid', bins = 50)
ax = plt.gca()
ax.set_xlabel('TPSA (angstrom)')
ax.set_ylabel('Number of Molecules')
ax.set_title("Number of Molecules in each TPSA (angstrom) Category", size=11.5, weight='bold') 
plt.show()
```

## Graph a Box Plot of TPSA for BBB+ and BBB-
Graph a box plot of the distribution of TPSA for the BBB+ group and the distribution of TPSA for the BBB- group

```python
'''Make a Box Plot of the TPSA Value, BBB+ and BBB-'''
fig, ax = plt.subplots(1, 2)
ax[0].boxplot(tpsa_positive, vert = False, showfliers = False)
ax[0].set_title("TPSA Distribution for BBB+ molecules", size=8, weight='bold')
ax[0].set_xlabel("TPSA (angstrom)")
ax[1].boxplot(tpsa_negative, vert = False, showfliers = False)
ax[1].set_title("TPSA Distribution for BBB- molecules", size=8, weight='bold')
ax[1].set_xlabel("TPSA (angstrom)")
plt.show()
```

## Graph a Box Plot of LogP Value for BBB+ and BBB- 
Graph a box plot of the distribution of logP for the BBB+ group and the distribution of logP for the BBB- group

```python
'''Make a Box Plot of the logP Value, BBB+ and BBB-'''
fig, ax = plt.subplots(1, 2)
ax[0].boxplot(logP_positive, vert = False, showfliers = False)
ax[0].set_title("logP Distribution for BBB+ molecules", size=10, weight='bold')
ax[0].set_xlabel("logP")
ax[1].boxplot(logP_negative, vert = False, showfliers = False)
ax[1].set_title("logP Distribution for BBB- molecules", size=10, weight='bold')
ax[1].set_xlabel("logP")
pos = ax[1].get_position() 
ax[1].set_position([pos.x0+0.07, pos.y0, pos.width, pos.height])
plt.show()
```

## Graph a Histogram of LogP Value and TPSA for combined dataset 
Graph a histogram of logP value for the entire data table and graph a histogram of TPSA for the entire data table

```python
'''Make a Histogram of LogP Value and TPSA for combined dataset'''
fig, ax = plt.subplots(1, 2)
bins = np.arange(min(logPs), max(logPs) + 1, 1)
ax[0].hist(logPs, edgecolor ='black', align='mid', bins = bins)
ax[0].set_xlabel('LogP Value')
ax[0].set_ylabel('Number of Molecules')
ax[0].set_title("Number of Molecules in each LogP Value Category", size=9, weight='bold') 
ax[1].hist(logPs, edgecolor ='black', align='mid', bins = 50)
ax[1].set_xlabel('TPSA (angstrom)')
ax[1].set_ylabel('Number of Molecules')
ax[1].set_title("Number of Molecules in each TPSA (angstrom) Category", size=9, weight='bold') 
pos = ax[1].get_position() 
ax[1].set_position([pos.x0+0.2, pos.y0, pos.width, pos.height])
plt.show()
```

## Graph a Histogram of LogP Value and TPSA individually for BBB+ and BBB-
Graph a histogram of logP value for the BBB permeable group, a histogram of logP value for the BBB nonpermeable group, a histogram of TPSA value for the BBB permeable group, and histogram of TPSA value for the BBB nonpermeable group.

```python
'''Make a Histogram of LogP Value and TPSA each with BBB+ and BBB-'''
fig, ax = plt.subplots(2, 2)
ax[0, 0].hist(logP_positive, edgecolor ='black', align='mid', bins = 20)
ax[0, 0].set_xlabel('LogP Value')
ax[0, 0].set_ylabel('Number of Molecules')
ax[1, 0].hist(tpsa_positive, edgecolor ='black', align='mid', bins = 30)
ax[1, 0].set_xlabel('TPSA (angstrom)')
ax[1, 0].set_ylabel('Number of Molecules')
ax[0, 1].hist(logP_negative, edgecolor ='black', align='mid', bins = 20)
ax[0, 1].set_xlabel('LogP Value')
ax[0, 1].set_ylabel('Number of Molecules')
ax[1, 1].hist(tpsa_negative, edgecolor ='black', align='mid', bins = 30)
ax[1, 1].set_xlabel('TPSA (angstrom)')
ax[1, 1].set_ylabel('Number of Molecules')
pos = ax[0,1].get_position() 
ax[0,1].set_position([pos.x0+0.1, pos.y0, pos.width, pos.height])
pos = ax[1,0].get_position() 
ax[1,0].set_position([pos.x0, pos.y0-0.1, pos.width, pos.height])
pos = ax[1,1].get_position() 
ax[1,1].set_position([pos.x0+0.1, pos.y0-0.1, pos.width, pos.height])
plt.show()
```

## Graph a Vertical Box Plot of TPSA and logP individually for BBB+ and BBB- 
Graph a vertical box plot of TPSA for the BBB permeable group and BBB nonpermeable group, where the two box plots share a combined axis. Graph a vertical box plot of logP for the BBB permeable group and BBB nonpermeable group, where the two box pltos share a combined axis.

```python
'''Make a Vertical Box Plot of the TPSA and logP, BBB+ and BBB- with combined axis'''
fig, ax = plt.subplots(1, 2)
ax[0].boxplot(tpsa_values, vert = True, showfliers = False)
ax[0].set_xticks([1, 2], ['BBB+', 'BBB-'])
ax[0].set_ylabel("TPSA (angstrom)")
ax[0].set_ylim(-50, 500)
ax[1].boxplot(logP_values, vert = True, showfliers = False)
ax[1].set_xticks([1, 2], ['BBB+', 'BBB-'])
ax[1].set_ylabel("LogP Value")
ax[1].set_ylim(-5, 10)
pos = ax[1].get_position() 
ax[1].set_position([pos.x0+0.1, pos.y0, pos.width, pos.height])
plt.show()
```
